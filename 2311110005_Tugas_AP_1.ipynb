```python
# Mengimpor library yang dibutuhkan
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
# Memasukan file dataset dan menampilkannya
file_path = "/content/CAD-Dataset.xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
# Dikarenakan dalam dataset terdapat data berupa '?' maka perlu diubah menjadi NaN agar missing value nya dapat terdeteksi semua
# Mengubah semua '?' menjadi NaN agar bisa dideteksi sebagai missing values
df.replace('?', np.nan, inplace=True)
```

```python
# Menampilkan jumlah missing value
df.isnull().sum()
```

```python
# Dikarenakan terdapat missing value pada data maka akan diisi dengan modus
for column in df.columns:
    if df[column].isnull().sum() > 0:  # Hanya memproses kolom yang memiliki missing values
        mode_series = df[column].mode()  # Mengambil mode (modus)

        if not mode_series.empty:  # Pastikan mode tidak kosong sebelum mengambil nilai
            mode_value = mode_series[0]  # Ambil nilai modus pertama
            df[column].fillna(mode_value, inplace=True)  # Mengisi missing values dengan modus
```

```python
# Dikarenakan pada kolom BMI tidak terdapat data maka kolom tersebut akan dihapus
df.drop(columns=['BMI'], inplace=True)
```

```python
df
```

```python
# Mengubah semua kolom kategorikal menjadi numerik 0 dan 1
kolom_kategorikal = [
    'Sex', 'DM', 'HTN', 'Current Smoker', 'EX-Smoker', 'FH', 'Obesity', 'CRF', 'CVA',
    'Airway disease', 'Thyroid Disease', 'ChF', 'DLP', 'Edema', 'Weak Peripheral Pulse',
    'Lung rales', 'Systolic Murmur', 'Diastolic Murmur', 'Dyspnea', 'Atypical',
    'Nonanginal', 'Exertional', 'LowTH Ang', 'LVH', 'Poor R Progression', 'VHD', 'Cath'
]

for col in kolom_kategorikal:
    if col in df.columns:
        unique_values = df[col].dropna().unique()

        if len(unique_values) == 2:  # Label Encoding
            # Mengkonversi menjadi 0 untuk N dan 1 untuk Y
            df[col] = df[col].map({unique_values[0]: 1, unique_values[1]: 0})

        elif len(unique_values) > 2:  # One-Hot Encoding
            df = pd.get_dummies(df, columns=[col], drop_first=True, dtype=int) # Specify dtype for one-hot columns

# Mengisi NaN dengan 0 sebelum konversi ke integer
df.fillna(0, inplace=True)

# Mengganti data non-numerik dengan 0
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0) # Handle errors, then fill NaN

# Konversi semua kolom boolean menjadi integer (0 dan 1)
df = df.astype(int)
```

```python
# Melakuakan normalisasi
from sklearn.preprocessing import MinMaxScaler

# Mengisi NaN dengan 0 agar tidak error saat normalisasi
df.fillna(0, inplace=True)

# Normalisasi dengan Min-Max Scaling (0 - 1)
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Menampilkan 5 baris pertama hasil normalisasi
print("\nData setelah normalisasi:\n", df_normalized.head())
```
